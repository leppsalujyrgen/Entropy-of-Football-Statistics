# ENTROPY OF FOOTBALL STATISTICS

**In this project we will be exploring football statistics and the effectivness of machine learning models to predict the outcome of matches. For this we will be employing two types of models: regression and classifier models.**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
statistics = pd.DataFrame()
for i in range (2016, 2019, 1):
    foldername = "Parsed_" + str(i) + "_" + str(i+1)[2:] + "/gwgames/"
    for j in range(1,39,1):
        filename = "gw" + str(j) + ".csv"
        statistics = statistics.append(pd.read_csv(foldername + filename))
```

```python
def encode_numeric_to_binary(dataframe, column_name, ranges, is_single=False):
    # Method generates binary columns from user defined ranges.
    #     dataframe : pandas dataframe, where the binary columns are put.
    #     column_name : the column that is transformed into binary columns.
    #     is_single : truth value that determines whether values in ranges array are interpreted as columns individually or in pairs.
    
    if is_single:
        for i in range(0,len(ranges)):
            binary_column_name = column_name + "_" + str(ranges[i])
            dataframe[binary_column_name] = np.where((dataframe[column_name] == ranges[i]), 1, 0)
    else:
        for i in range(0,len(ranges)):
            binary_column_name = column_name + "_" + str(ranges[i][0]) + "_" + str(ranges[i][1])
            dataframe[binary_column_name] = np.where((ranges[i][0] <= dataframe[column_name])&(dataframe[column_name] <= ranges[i][1]), 1, 0)
            

    return dataframe.drop(columns=[column_name], axis=1)
    
    
statistics_binary = statistics.copy(deep=True)
# Dropping useless columns.
statistics_binary = statistics_binary.drop(columns=['fixture', 'round', 'minutes'], axis=1)

# Transforming numeric columns into binary columns.
# Encoding attempted passes
statistics_binary = encode_numeric_to_binary(statistics_binary, 'attempted_passes', [(0,249), (250,539), (540,1015)])
# Encoding big_chances_created
statistics_binary = encode_numeric_to_binary(statistics_binary, 'big_chances_created', [0,1,2,3,4,5,6,7,8], 1)
# Encoding big_chances_missed0
statistics_binary = encode_numeric_to_binary(statistics_binary, 'big_chances_missed', [0,1,2,3,4,5,6], 1)
# Encoding clearances_blocks_interceptions
statistics_binary = encode_numeric_to_binary(statistics_binary, 'clearances_blocks_interceptions', [(10,19), (20,39), (40,49), (50,59), (60,69), (70,79), (80,89), (90,99)])
# Encoding completed_passes
statistics_binary = encode_numeric_to_binary(statistics_binary, 'completed_passes', [(0,249), (250,539), (540,1015)])
# Encoding dribbles
statistics_binary = encode_numeric_to_binary(statistics_binary, 'dribbles', [(0,0),(0,1),(2,5),(6,12),(13,28)])
# Encoding errors_leading_to_goal_attempt
statistics_binary = encode_numeric_to_binary(statistics_binary, 'errors_leading_to_goal_attempt', [1,2,3,4], 1)
# Encoding fouls
statistics_binary = encode_numeric_to_binary(statistics_binary, 'fouls', [(0,1),(2,7),(8,13),(14,26)])
# Encoding key_passes
statistics_binary = encode_numeric_to_binary(statistics_binary, 'key_passes', [(0,5), (6,12), (13,32)])
# Encoding offside
statistics_binary = encode_numeric_to_binary(statistics_binary, 'offside', [(0,1), (2,3), (4,10)])
# Encoding open_play_crosses
statistics_binary = encode_numeric_to_binary(statistics_binary, 'open_play_crosses', [(0,0), (1,4), (5,10)])
# Encoding opponent_team
statistics_binary = encode_numeric_to_binary(statistics_binary, 'opponent_team', list(range(20)), 1)
# Encoding penalties_conceded
statistics_binary = encode_numeric_to_binary(statistics_binary, 'penalties_conceded', [1,2], 1)
# Encoding penalties_missed
statistics_binary = encode_numeric_to_binary(statistics_binary, 'penalties_missed', [1,2], 1)
# Encoding penalties_saved
statistics_binary = encode_numeric_to_binary(statistics_binary, 'penalties_saved', [1,2], 1)
# Encoding recoveries
statistics_binary = encode_numeric_to_binary(statistics_binary, 'recoveries', [(20,54), (55,66), (67,94)])
# Encoding red_cards
statistics_binary = encode_numeric_to_binary(statistics_binary, 'red_cards', [1, 2], 1)
# Encoding saves
statistics_binary = encode_numeric_to_binary(statistics_binary, 'saves', [(0,1), (2,5), (6,14)])
# Encoding tackled
statistics_binary = encode_numeric_to_binary(statistics_binary, 'tackled', [(0,1), (2,11), (12,19), (20,38)])
# Encoding tackles
statistics_binary = encode_numeric_to_binary(statistics_binary, 'tackles', [(0,1),(2,4),(5,10),(11,26)])
# Encoding target_missed
statistics_binary = encode_numeric_to_binary(statistics_binary, 'target_missed', [(0,1),(2,6),(7,16)])
# Encoding team_score
statistics_binary = encode_numeric_to_binary(statistics_binary, 'team_score', [0,1,2,3,4,5,6,7],1)
# Encoding yellow_cards
statistics_binary = encode_numeric_to_binary(statistics_binary, 'yellow_cards', [(0,0), (1,2), (3,16)])
# Encoding shots_on_target
statistics_binary = encode_numeric_to_binary(statistics_binary, 'shots_on_target', [(0,0), (1,2), (3,9), (10,17)])
# Encoding name
for name in list(statistics_binary["name"].unique()):
    statistics_binary[name] = np.where((statistics_binary["name"] == name), 1, 0)
statistics_binary = statistics_binary.drop(columns=["name"], axis=1)
statistics_binary
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Finding frequent patterns in dataset
pd.set_option('display.max_rows', 1200)
pd.set_option('display.max_colwidth', 200)
frequent_itemset = apriori(statistics_binary, min_support=0.45, use_colnames=True)
frequent_itemset

# Finding rules
mlx_rules = association_rules(frequent_itemset,metric='confidence',min_threshold=0.3)
mlx_rules
```
