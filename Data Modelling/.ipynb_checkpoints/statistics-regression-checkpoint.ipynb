```python
statistics = pd.DataFrame()
for i in range (2016, 2019, 1):
    foldername = "Parsed_" + str(i) + "_" + str(i+1)[2:] + "/gwgames/"
    for j in range(1,39,1):
        filename = "gw" + str(j) + ".csv"
        statistics = statistics.append(pd.read_csv(foldername + filename))
```

```python
def encode_numeric_to_binary(dataframe, column_name, ranges, is_single=False):
    
    for i in range(1,len(ranges)):
        if is_single:
            binary_column_name = column_name + "_" + str(ranges[i-1])
        else:
            binary_column_name = column_name + "_" + str(ranges[i-1]) + "_" + str(ranges[i])
                
        dataframe[binary_column_name] = np.where((ranges[i-1] < dataframe[column_name])&(dataframe[column_name] <= ranges[i]), 1, 0)


   
    return dataframe.drop(columns=[column_name], axis=1)
    
    
# Transforming numeric columns into binary columns.
statistics_binary = statistics.copy(deep=True)


# Encoding attempted passes
statistics_binary = encode_numeric_to_binary(statistics_binary, 'attempted_passes', [0,200,400,600,800,1000,1200])
# Encoding big_chances_created
statistics_binary = encode_numeric_to_binary(statistics_binary, 'big_chances_created', list(range(statistics_binary['big_chances_created'].max()+1)), 1)
# Encoding big_chances_missed
statistics_binary = encode_numeric_to_binary(statistics_binary, 'big_chances_missed', list(range(statistics_binary['big_chances_missed'].max()+1)), 1)
# Encoding clearances_blocks_interceptions
statistics_binary = encode_numeric_to_binary(statistics_binary, 'clearances_blocks_interceptions', list(range(0, 101, 10)))
# Encoding completed_passes
statistics_binary
# Encoding dribbles
# Encoding errors_leading_to_goal_attempt
# Encoding fixture
# Encoding fouls
# Encoding key_passes
# Encoding minutes
# Encoding offside
# Encoding open_play_crosses
# Encoding opponent_team
# Encoding penalties_conceded
# Encoding penalties_missed
# Encoding penalties_saved
# Encoding recoveries
# Encoding red_cards
# Encoding round
# Encoding saves
# Encoding tackled
# Encoding tackles
# Encoding target_missed
# Encoding team_score
# Encoding was_home
# Encoding yellow_cards
# Encoding shots_on_target
```
